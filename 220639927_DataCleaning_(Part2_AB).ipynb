```python
#Importing the required libraries
import numpy as np
import pandas as pd
```

```python
#Loading the dataframes from 2006 and 2007
df06 = pd.read_csv('C:\\Users\\bhagy\\Programming CourseWork\\Python\\2006.csv')
df07 = pd.read_csv('C:\\Users\\bhagy\\Programming CourseWork\\Python\\2007.csv')
```

```python
#Combining the two dataframes into one dataframe
df = pd.concat([df06, df07])
df.head()
```

```python
#Identifying the null values in the dataframe
pd.isnull(df).sum(axis=0)
```

```python
#Removing the column 'CancellationCode'
df=df.drop(columns=['CancellationCode'])
df.head()
```

```python
#Removing the rows containing null values
df = df.dropna(axis=0)
```

```python
#Checking for null values after removal
#If the null values from the column 'CancellationCode' were to be omitted, there would be insufficient data
pd.isnull(df).sum(axis=0)
```

```python
#Checking for unreasonable values/outliers by analysing the minimun and maximum values of each column
for col_name in df.columns:
    print(f'Colname: {col_name}, min: {df[col_name].min()}, max: {df[col_name].max()}')
```

```python
#All data for 'Cancelled' and 'Diverted' flights have been removed, but they won't be needed for the first two parts of the question.
#The negative values existing in the columns 'ArrDelay' and 'DepDelay' are assumed to be a result of early flights.
#The maximum values in 'DepTime' and 'AirTime' have exceeded 2400, which are considered as unreasonable values.
#The maximum values are therefore rectified to 2359.
df = df[(df["DepTime"]<2400) & (df["ArrTime"] <2400)]
```

```python
#Rechecking after rectification
for col_name in df.columns:
    print(f'Colname: {col_name}, min: {df[col_name].min()}, max: {df[col_name].max()}')
```

```python
#Saving the cleaned dataframe for Flight Analysis
df.to_csv('Cleaned_Dataframe_for_Flight_Analysis.csv')
```
