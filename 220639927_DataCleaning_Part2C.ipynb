# Data Cleaning for Flight Analysis

For Part2 QC

(The previously cleaned dataset had all 0 values for the 'diverted' column)

```python
#Importing numpy and pandas libraries
import numpy as np
import pandas as pd
```

```python
#Importing the dataframes from years 2006 and 2007
df1 = pd.read_csv('C:\\Users\\bhagy\\Programming CourseWork\\Python\\2006.csv', encoding = 'latin1')
df2 = pd.read_csv('C:\\Users\\bhagy\\Programming CourseWork\\Python\\2007.csv', encoding = 'latin1')
```

```python
#Combining the dataframes into one dataframe
df = pd.concat([df1, df2])
df
```

```python
#Identifying the null values in the dataframe
pd.isnull(df).sum(axis=0)
```

```python
#Removing the columns containing null values
df = df.dropna(axis=1)
#Rechecking for null after removing columns
pd.isnull(df).sum(axis=0)
```

```python
#Inspecting value counts for 'Diverted'
df['Diverted'].value_counts()
```

```python
#Checking for unreasonable values by analysing the minimun and maximum values of each column
for col_name in df.columns:
    print(f'Colname: {col_name}, min: {df[col_name].min()}, max: {df[col_name].max()}')
```

```python
#Rectifying the maximum value for 'CRSDepTime'
df = df[(df['CRSArrTime']<2400)]
```

```python
#Rechecking for unreasonable values by analysing the minimun and maximum values of each column
for col_name in df.columns:
    print(f'Colname: {col_name}, min: {df[col_name].min()}, max: {df[col_name].max()}')
```

```python
#Viewing the dataset
df.head()
```

```python
#Saving the cleaned dataframe for Flight Analysis
df.to_csv('Cleaned_Dataframe_Part2QC_for_Flight_Analysis.csv')
```
