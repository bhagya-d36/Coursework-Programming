```python
#Importing the required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Loading the cleaned dataframes
df = pd.read_csv('C:\\Users\\bhagy\\Programming CourseWork\\Python\\Cleaned_Dataframe_for_Flight_Analysis.csv')
df_planedata = pd.read_csv('C:\\Users\\bhagy\\Programming CourseWork\\Python\\Plane_Data_Cleaned.csv')
```

```python
#Viewing the cleaned dataframe
df.head()
```

```python
#Dropping 'Unnamed: 0' column
df=df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
#Viewing the plane data
df_planedata
```

```python
#Dropping 'Unnamed: 0' column in plane data
df_planedata=df_planedata.drop(columns=['Unnamed: 0'])
df_planedata
```

```python
#Checking for null values in df
pd.isnull(df).sum()
```

```python
#Checking for null values in plane-data
pd.isnull(df_planedata).sum()
```

```python
#Obtaining only the neccessary columns and creating a simpler dataframe for analysis
df1=df[['TailNum', 'Year', 'Month', 'DayofMonth', 'DayOfWeek','DepDelay','ArrDelay','DepTime']]
df1.head()
```

```python
#Renaming columns in 'df_planedata' for merging purposes
df_planedata.rename(columns={'tailnum':'TailNum', 'year': 'YearOfManufacture'},inplace=True)
df_planedata
```

```python
#Merging the 2 datasets on TailNum
merged_df=pd.merge(df1,df_planedata,on='TailNum')
merged_df.head()
```

```python
#Checking for any null values
pd.isnull(merged_df).sum()
```

```python
#Adding a new column TotalDelay and replacing all the negative values in that column with 0
merged_df['TotalDelay']=merged_df.ArrDelay+merged_df.DepDelay
merged_df['TotalDelay'] = merged_df['TotalDelay'].clip(lower=0)
merged_df.head()
```

```python
#Checking for unreasonable values in the Year of Manufacture column
merged_df['YearOfManufacture'].value_counts()
```

```python
#The entry '0' has to be removed
merged_df = merged_df[merged_df['YearOfManufacture'] != 0]
merged_df['YearOfManufacture'].value_counts()
```

```python
# Seperating New Airplanes and Old Airplanes
# Assumption : Planes manufactured before the year 1983 are considered old airplanes.
merged_df['YearOfManufacture'] = merged_df['YearOfManufacture'].astype(int)
def age(x):
    if x <1982:
        return 'Old'
    else:
        return 'New'
merged_df['PlaneCondition']= merged_df['YearOfManufacture'].apply(lambda x:age(x))
merged_df.head()
```

```python
#Creating a new column called 'PlaneAge'
merged_df['PlaneAge']=merged_df['Year'].astype(int)-merged_df['YearOfManufacture'].astype(int)
merged_df.head()
```

```python
#Checking for unreasonable entries in 'PlaneAge'
merged_df['PlaneAge'].value_counts()
```

```python
#Removing the entry '-1' in 'PlaneAge'
merged_df = merged_df[(merged_df['PlaneAge'] != -1)]
#Rechecking
merged_df['PlaneAge'].value_counts()
```

```python
#Splitting the dataframe into the 2 years for the analysis
df_2006 = merged_df[merged_df['Year'] == 2006]
df_2007 = merged_df[merged_df['Year'] == 2007]
```

```python
#Simplifying the dataframe by taking only the required columns
df_2006=df_2006[['Year','YearOfManufacture','TotalDelay','PlaneCondition','PlaneAge']]
df_2006
```

```python
#Simplifying the dataframe by taking only the required columns
df_2007=df_2007[['Year','YearOfManufacture','TotalDelay','PlaneCondition','PlaneAge']]
df_2007
```

---

# Analysing the year 2006

```python
#Creating a dataframe obtaining the mean total delay for every year of manufacture for planes.
AvgDelay_MfgYear_06=df_2006[['YearOfManufacture','TotalDelay']].groupby(['YearOfManufacture']).mean().reset_index(drop=False)
AvgDelay_MfgYear_06
```

```python
#Obtaining a new average delay with respect to Old and New airplanes
PlaneCondition_Delay2006 = df_2006.groupby('PlaneCondition')['TotalDelay'].mean().reset_index()
PlaneCondition_Delay2006 = PlaneCondition_Delay2006.rename(columns={'TotalDelay':'AvgTotalDelay'})
PlaneCondition_Delay2006
```

In 2006, the average total delay is greater for newer planes in comparsion to older planes.

```python
#Obtaining a new dataframe with Plane Age and Mean Total Delay
AvgDelay_Age06=df_2006[['PlaneAge','TotalDelay']].groupby('PlaneAge').mean().reset_index()
AvgDelay_Age06 = AvgDelay_Age06.rename(columns={'TotalDelay':'MeanTotalDelay'})
AvgDelay_Age06
```

```python
#Correlation Coefficient
AvgDelay_Age06['PlaneAge'].corr(AvgDelay_Age06['MeanTotalDelay'])
```

The correlation coefficient indicates a weak negative correlation

# Analysing the year 2007

```python
#Creating a dataframe obtaining the mean total delay for every year of manufacture for planes.
AvgDelay_MfgYear_07=df_2007[['YearOfManufacture','TotalDelay']].groupby(['YearOfManufacture']).mean().reset_index(drop=False)
AvgDelay_MfgYear_07
```

```python
#Obtaining a new average delay with respect to Old and New airplanes
PlaneCondition_Delay2007 = df_2007.groupby('PlaneCondition')['TotalDelay'].mean().reset_index()
PlaneCondition_Delay2007 = PlaneCondition_Delay2007.rename(columns={'TotalDelay':'AvgTotalDelay'})
PlaneCondition_Delay2007
```

In 2007, the average total delay is greater for older planes in comparsion to newer planes.

```python
#Obtaining a new dataframe with Plane Age and Mean Total Delay
AvgDelay_Age07=df_2007[['PlaneAge','TotalDelay']].groupby('PlaneAge').mean().reset_index()
AvgDelay_Age07 = AvgDelay_Age07.rename(columns={'TotalDelay':'MeanTotalDelay'})
AvgDelay_Age07
```

```python
#Correlation Coefficient
AvgDelay_Age07['PlaneAge'].corr(AvgDelay_Age07['MeanTotalDelay'])
```

The correlation coefficient indicates a moderate positive correlation

# Visualisations

```python
#Visualising both years on the same figure to obtain the mean total delay based on plane condition
#Differentiating 2006 and 2007 by creating a new merged dataframe for visualisation purposes
PlaneCondition_Delay2006['Year'] = '2006'
PlaneCondition_Delay2007['Year'] = '2007'
#Merging the dataframes after seperating the 2 years.
PlaneCondition_Delay0607 = pd.concat([PlaneCondition_Delay2006, PlaneCondition_Delay2007])
#Creating the figure (barplot)
plt.figure(figsize=(6, 6))
sns.barplot(x='Year', y='AvgTotalDelay', hue='PlaneCondition', data=PlaneCondition_Delay0607, palette={'New':'darkcyan', 'Old':'darkslategray'})
plt.xlabel('Year', fontsize = 15)
plt.ylabel('Average Total Delay', fontsize = 15)
plt.legend(title='Plane Condition',title_fontsize = 12, bbox_to_anchor=(1, 1), loc='upper left', fontsize = 10)
plt.show()
```

```python
#Visualising both years on the same figure to obtain the mean total delay based on Plane Age
plt.figure(figsize=(10, 8))
sns.set_style('white')
#Regression plot for 2007
sns.regplot(x='PlaneAge', y='MeanTotalDelay', data=AvgDelay_Age07, \
            scatter_kws={'s': 75}, color = 'crimson', line_kws={'color':'crimson'}, \
            label='2007')
#Regression plot for 2006
sns.regplot(x='PlaneAge', y='MeanTotalDelay', data=AvgDelay_Age06, \
            scatter_kws={'s': 75}, color = 'darkcyan',line_kws={'color': 'darkcyan'}, \
            label='2006')
#Customizing the figure
plt.title('Average Delay vs Age of Plane', fontsize = 25)
plt.xlabel('Age of Plane', fontsize = 20)
plt.xticks(fontsize=15)
plt.ylabel('Average Delay', fontsize = 20)
plt.yticks(fontsize=15)
plt.legend(loc = 'upper left', fontsize = 15)
plt.tight_layout()
plt.show()
#Displaying the correlation coefficients values for each year
correlation_coefficient = AvgDelay_Age06['PlaneAge'].corr(AvgDelay_Age06['MeanTotalDelay'])
print('Correlation coefficient (2006):', correlation_coefficient)
correlation_coefficient = AvgDelay_Age07['PlaneAge'].corr(AvgDelay_Age07['MeanTotalDelay'])
print('Correlation coefficient (2007):', correlation_coefficient)
```

An overall increase in the average delay with age of plane is evident in 2007 while a decrease is seen in 2006.
